## Tune hyper-parameter

Here, my composition varies 0 to 1 in x, y, z direction therefore, [0.,1.] is my boundary for the dataset
Setting of a hyper parameter, the thumb rule is to take the differece of max and min and perfrom square of it.

```python
index_set_bounds = np.array([[0., 1.], [0., 1.], [0., 1.]])
#hyperparameter_bounds = np.array([[0.0001, 1e7], [0.0001, 100]])
hyperparameter_bounds = np.array([[0.0001,1e7],[0.0001,100], [7,14], [10, 14], [10,17], [2500, 3000], [2000, 2600], [3000, 3600]])
likelihood_optimization_method = "global"
likelihood_optimization_max_iter = 120
likelihood_optimization_pop_size = 100
likelihood_optimization_tolerance = 0.000001
hps_guess = np.ones((8))
```

## Kernel selection
I am using matern kernel for my dataset

```python
def kernel_l2_single_task(x1, x2, hyperparameters, obj):
    """
    x1: 2d numpy array of points
    x2: 2d numpy array of points
    obj: object containing kernel definition
    Return:
    -------
    Kernel Matrix
    """
    hps = hyperparameters
    distance_matrix = np.zeros((len(x1), len(x2)))
    for i in range(len(x1[0]) - 1):
        distance_matrix += abs(np.subtract.outer(x1[:, i], x2[:, i]) /
                               hps[1])**2
    distance_matrix = np.sqrt(distance_matrix)
    return hps[0] * obj.matern_kernel_diff1(distance_matrix, 1)
```

# Mean function : Klemens model

## Hyper-parameter tuning in Physics based model
Model and fearization files came from Klemens model, here I am directly implementing those model and features

```python
#%%
Ta_m = 	180.95  #row1
V_m =	50.942  #row2
Nb_m =	92.906  #row3
Fe_m =	55.845
Sb_m = 	121.76

n = 3  #number of atoms total

#average speed of light
#Ta_speed = 2907
#V_speed = 2374
#Nb_speed = 3473
#Nb_speed = 3052


#Lattice thermal conductivity
#Ta_LTC = 8.8 
#V_LTC  = 12.2
#Nb_LTC = 14 


#average volume per atom
Ta_aV = 1.74257E-29
V_aV  = 1.62392E-29
Nb_aV = 1.77223E-29


const1 = 6 * (np.pi)**5
const2 = 2*1.380649e-23
#print(const2)

def mass (X, Y):
    f = (X*Ta_m) + (Y*V_m) + (1-X-Y)*Nb_m 
    return f

def speed_light (X1, Y1, hps):
    Ta_speed = hps[5]
    V_speed  = hps[6]
    Nb_speed = hps[7] 
    f1 = (X1*Ta_speed) + (Y1*V_speed) + (1-X1-Y1)*Nb_speed
    return f1

def LTC_p (X2, Y2, hps):
    Ta_LTC = hps[2]
    V_LTC  = hps[3]
    Nb_LTC = hps[4] 
    f2 = (X2*Ta_LTC) + (Y2*V_LTC) + (1-X2-Y2)*Nb_LTC
    return f2

def ave_volume (X3, Y3):
    f3 = (X3*Ta_aV) + (Y3*V_aV) + (1-X3-Y3)*Nb_aV
    return f3

def get_gamma (x,y):
    mass_M = mass (x, y)
    #calulation of gamma
    M_total = mass_M + Fe_m + Sb_m
    M_bar = M_total/n
    A1 = x*(Ta_m - mass_M)**2
    A2 = y*(V_m - mass_M)**2
    A3 = (1.0-x-y)*(Nb_m - mass_M)**2
    average_mass_variance = (A1+A2+A3)/n
    gamma = average_mass_variance / (M_bar)**2
    return gamma

def get_u (x,y,hps):
    gamma = get_gamma (x,y)
    speed_l = speed_light (x, y, hps)
    #average lattice thermal conductivity 
    LTC_pure = LTC_p (x,y,hps) 
    #average volume per atom calculation
    ave_volume_atom = ave_volume(x,y)
    #calculations
    numerator = ((const1)*(ave_volume_atom)**2)**(1/3)
    denominator = const2 * speed_l 
    RHS = (gamma*LTC_pure*numerator)/denominator

    def func(U):
        return [U[0] * U[1] - RHS,
                U[0] - np.tan(U[1])]
    if x==0:
        if y == 0 :
            root = fsolve(func, [0.1, 0.1])   
        else:
            root = fsolve(func, [RHS, math.atan(RHS)])   
    else:
        root = fsolve(func, [RHS/math.atan(RHS), math.atan(RHS)])
    valid = np.isclose(func(root), [0.0, 0.0])
    #print(str(valid))
    return root

def get_kl (x, y,hyperparameters):
    if x == 1:
        x = 0.99999999
    if y ==1:
        y = 0.99999999
    solution =  get_u (x,y,hyperparameters)  
    k0 = LTC_p (x,y,hyperparameters) 
    kl = k0*solution[1]/solution[0]
    #print(str(solution[0])) 
    #print(str(kl)) 
    return kl

#get_kl(0, 1)
```

## Mean function : Incorporate Matminer in mean function

I am using the model for FeSb(TaFeSb, VFeSb and NbFeSb) family and incorparting that to the GPR mean function

```python
def get_kl_array_func(gp_obj, X, hyperparameters):
    #global x
    kl = np.empty((len(X)))
    for i in range(len(X)):
        kl[i] = get_kl (X[i, 0], X[i, 1],hyperparameters)
    return kl
```

```python
%%time
gp.tell(x, y)
gp.init_gp(hps_guess,
           gp_kernel_function=kernel_l2_single_task,
           gp_mean_function=get_kl_array_func)
gp.train_gp(hyperparameter_bounds)
```

# Step 2: Visilization (Plot ternary)

## Create the quternary plots for the datasets
Scale should be silimar as the ternary plot

```python
from ternary.helpers import simplex_iterator, normalize
P = []
Q = []
R = []
t = []
for e, f, g in simplex_iterator(scale=20, boundary=True):
    p, q, r = normalize([e, f, g])
    t.append((p, q, r))
    P.append(p)
    Q.append(q)
    R.append(r)
```

## Prediction (posterior_mean) from feature-based GPR

```python
counter = 0
l = np.empty((len(P), 4))
for i in range(len(P)):
    res = gp.gp.posterior_mean(np.array([[P[i], Q[i], R[i]]]))
    print(i)
    print(res)
    l[counter, 0] = P[i]
    l[counter, 1] = Q[i]
    l[counter, 2] = R[i]
    l[counter, 3] = res["f(x)"]  #no obj
    counter += 1
```

### Remove the point and comapare the model (LOOCV) 

```python
pred_loocv = []
exp = []
error = 0
error_array = np.empty((len(x)))
for index in range(0,len(x),1):
    #print("<<<<<<<<<<<<<<<<<<<<<<< " + str(index) + " >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
    #print("<<<<<<<<<<<<<<<<<<<<<<< " + str(y_cv) + " >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
    x_n = np.array([x[index]])
    y_n = np.array(y[index])
    x_cv = np.delete(x, (index), axis=0)
    y_cv = np.delete(y, (index), axis=0)
    
    gp1.tell(x_cv, y_cv)
    gp1.init_gp(hps_guess,
           gp_kernel_function=kernel_l2_single_task,
           gp_mean_function=get_kl_array_func)
        
    res = gp1.gp.posterior_mean(x_n)["f(x)"][0]
    #print(res,y_n)
    pred_loocv.append(res)
    exp.append(y_n)
    error_array[index] = abs(res-y_n[0])
    error += abs(res-y_n[0])
error = error/len(x)
print("average error: ", error)
print(error_array)
#print(exp)
#print(pred_loocv)
```
